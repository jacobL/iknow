```python
# step 2. 將 "Sheet標準題" 解析到 standand_question

connSQE = pymysql.connect( host = '10.55.52.98' , port=33060 , user = 'root' , passwd = "1234" , db = 'sqe')
curSQE = connSQE.cursor()

# step 2.1 qs_standand_question

sheet_raw = data.sheet_by_name('1.1標準題')
print("SQE標準題 row num:", sheet_raw.nrows,",col num:", sheet_raw.ncols) 
curSQE.execute("delete from sqe_standand_question")
for row in range(1, sheet_raw.nrows):   
    Answer_NO = sheet_raw.cell(row,0).value
    System = sheet_raw.cell(row,1).value
    Question = sheet_raw.cell(row,2).value
    category_main = sheet_raw.cell(row,3).value
    category_sub = sheet_raw.cell(row,4).value
    Answer = sheet_raw.cell(row,5).value
    Keyword = sheet_raw.cell(row,6).value
    SOP_chapter = sheet_raw.cell(row,7).value
    SOP_No = sheet_raw.cell(row,8).value.splitlines()[0].split()[0] if sheet_raw.cell(row,8).value !='' else '' 
    URL = sheet_raw.cell(row,9).value
    Enabled = 'True' #  sheet_raw.cell(row,10).value
    #Creator = sheet_raw.cell(row,11).value
    Create_Date = sheet_raw.cell(row,12).value     
    print(row,'. ',Answer_NO,',SOP_No:',SOP_No) 
    curSQE.execute("insert into sqe_standand_question(Answer_NO,System,Question,category_main,category_sub,Answer,Keyword,\
    SOP_chapter,SOP_No,URL,Enabled,Creator,Create_Date)values(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,CONVERT_TZ(NOW(),'+00:00','+8:00'))",\
    (Answer_NO,System,Question,category_main,category_sub,Answer,Keyword,SOP_chapter,SOP_No,URL,Enabled,Creator))
curSQE.execute("commit")  

connSQE.close()
```

```python
# step 3. 將 "Sheet相似題" 解析到 sim_question
connSQE = pymysql.connect( host = '10.55.52.98' , port=33060 , user = 'root' , passwd = "1234" , db = 'sqe')
curSQE = connSQE.cursor()
# step 3.1 qs_sim_question

sheet_raw = data.sheet_by_name('1.2 相似題')
print("SQE各系統問答資料庫 相似題 row num:", sheet_raw.nrows,",col num:", sheet_raw.ncols) 
curSQE.execute("delete from sqe_sim_question")
 
for row in range(1, sheet_raw.nrows):
    Answer_NO = sheet_raw.cell(row,0).value
    Question = sheet_raw.cell(row,1).value
    print(row,' ',Answer_NO,' ',Question) 
    
    curSQE.execute("insert ignore into sqe_sim_question(Answer_NO,Question,Creator,Create_Date)values(%s,%s,%s,CONVERT_TZ(NOW(),'+00:00','+8:00'))",\
    (Answer_NO,Question,Creator))
curSQE.execute("commit") 
connSQE.close()      
```

```python
# step 5. 將 sqe_standand_question裡的Keyword解析後，塞入sqe_jiebadict

import re  
connSQE = pymysql.connect( host = '10.55.52.98' , port=33060 , user = 'root' , passwd = "1234" , db = 'sqe')
curSQE = connSQE.cursor()
curSQE2 = connSQE.cursor()
curSQE.execute("delete from sqe_jiebadict")
curSQE.execute("SELECT Keyword FROM sqe_standand_question")

for Key in curSQE :
    KeywordList = re.split('，|；',Key[0])  
    for i in range(0,len(KeywordList)) :
        if len(KeywordList[i]) > 1 :
            curSQE2.execute("insert ignore into sqe_jiebadict(jiebaword,Creator,Create_Date)values(%s,%s,CONVERT_TZ(NOW(),'+00:00','+8:00'))",(KeywordList[i],Creator))
curSQE2.execute("commit") 
connSQE.close()      
```
